```python
# Import Key Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import time
from sklearn.linear_model import LinearRegression


from jupyterthemes import jtplot
jtplot.style(theme='monokai', context='notebook', ticks=True, grid=False) 
# setting the style of the notebook to be monokai theme  
# this line of code is important to ensure that we are able to see the x and y axes clearly
# If you don't run this code line, you will notice that the xlabel and ylabel on any plot is black on black and it will be hard to see them. 

# Data
x = np.array([56, 49, 72, 38, 63, 47])  # Age
y = np.array([147, 145, 160, 115, 130, 128])  # Blood Pressure

# Scatter plot
plt.scatter(x, y, color='blue', label='Data points')

# Fit regression line (y = m*x + b)
m, b = np.polyfit(x, y, 1)  # slope m and intercept b
y_pred = m * x + b

# Plot regression line
plt.plot(x, y_pred, color='red', label='Regression line')

# Labes and title
plt.xlabel('Age in Years (X)')
plt.ylabel('Blood Pressure (Y)')
plt.title('Age vs Blood Pressure')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Data
x = np.array([56, 49, 72, 38, 63, 47])          # Age
y = np.array([147, 145, 160, 115, 130, 128])    # Blood Pressure

# Best fit line (least squares)
m_best, b_best = np.polyfit(x, y, 1)

# x range for drawing lines
x_line = np.linspace(min(x) - 5, max(x) + 5, 100)

# Best fit line
y_best = m_best * x_line + b_best

# Some other candidate lines (manual slopes/intercepts)
y_line1 = (m_best + 0.5) * x_line + (b_best - 10)
y_line2 = (m_best - 0.5) * x_line + (b_best + 10)
y_line3 = m_best * x_line + (b_best + 20)

# Plot data
plt.scatter(x, y, color='black', label='Data points')

# Plot lines
plt.plot(x_line, y_best, color='red', linewidth=2, label='Best fit line')
plt.plot(x_line, y_line1, color='blue', linestyle='--', label='Candidate line 1')
plt.plot(x_line, y_line2, color='green', linestyle='--', label='Candidate line 2')
plt.plot(x_line, y_line3, color='purple', linestyle='--', label='Candidate line 3')

plt.xlabel('Age in Years (X)')
plt.ylabel('Blood Pressure (Y)')
plt.title('Many possible lines vs best fit')
plt.legend()
plt.grid(True)
plt.show()
```

# prediact using scikit learn libaray


```python
from sklearn.linear_model import LinearRegression
# Data
x = np.array([56, 49, 72, 38, 63, 47])          # Age
y = np.array([147, 145, 160, 115, 130, 128])    # Blood Pressure

# scikit-learn expects X as 2D array: shape (n_samples, n_features)
X = x.reshape(-1, 1)

# 1) Create and fit the model
model = LinearRegression()
model.fit(X, y)

# 2) Get slope and intercept (should be close to np.polyfit)
print("Slope (b1) from sklearn      :", model.coef_[0])
print("Intercept (b0) from sklearn  :", model.intercept_)

# 3) Predict blood pressure for the same ages
y_pred = model.predict(X)
print("Predicted BP for given ages:", y_pred)

# 4) Predict for a new age, e.g. 60 years
new_age = np.array([[60]])
new_bp_pred = model.predict(new_age)
print("Predicted BP for age 60:", new_bp_pred[0])
```

```python
# Data
x = np.array([56, 49, 72, 38, 63, 47])
y = np.array([147, 145, 160, 115, 130, 128])
X = x.reshape(-1, 1)   # 2D shape for sklearn

# Number of repetitions to get a stable timing
N = 10000

print("=== Timing np.polyfit ===")
start = time.time()
for _ in range(N):
    m, b = np.polyfit(x, y, 1)
end = time.time()
polyfit_time = end - start
print("Total time for np.polyfit:", polyfit_time, "seconds")

print("\n=== Timing sklearn LinearRegression ===")
start = time.time()
for _ in range(N):
    model = LinearRegression()
    model.fit(X, y)
end = time.time()
sklearn_time = end - start
print("Total time for LinearRegression:", sklearn_time, "seconds")

print("\nRatio (sklearn / polyfit):", sklearn_time / polyfit_time)
```

```python
# Data
x = np.array([56, 49, 72, 38, 63, 47])          # Age
y = np.array([147, 145, 160, 115, 130, 128])    # Blood Pressure

# Reshape X for sklearn
X = x.reshape(-1, 1)

# Fit model
model = LinearRegression()
model.fit(X, y)

# Create smooth line for plotting
x_line = np.linspace(x.min() - 2, x.max() + 2, 100).reshape(-1, 1)
y_line = model.predict(x_line)

# Plot data points
plt.scatter(x, y, color='blue', label='Data points')

# Plot regression line
plt.plot(x_line, y_line, color='red', label='Sklearn regression line')

plt.xlabel('Age in Years (X)')
plt.ylabel('Blood Pressure (Y)')
plt.title('Age vs Blood Pressure (Sklearn Linear Regression)')
plt.legend()
plt.grid(True)
plt.show()
```
